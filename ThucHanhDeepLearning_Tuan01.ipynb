```python
# @title
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu
path = 'https://raw.githubusercontent.com/realRandomPL/ThucHanhDeepLearning/refs/heads/main/Data/titanic_disaster.csv'
titanicDisaster = pd.read_csv(path)

# Xử lý cột Name: Tách thành firstName và secondName, sau đó xóa cột Name
titanicDisaster[['firstName', 'secondName']] = titanicDisaster['Name'].str.split(',', n=1, expand=True)
titanicDisaster = titanicDisaster.drop('Name', axis=1)

# Rút gọn dữ liệu cột Sex
titanicDisaster['Sex'] = titanicDisaster['Sex'].replace({'male': 'M', 'female': 'F'})

# Xử lý thiếu dữ liệu Age bằng trung bình theo hạng vé (Pclass)
# Tính trung bình tuổi của từng hạng vé
age_means = titanicDisaster.groupby('Pclass')['Age'].mean()
# Hàm điền tuổi theo hạng vé
def fill_age(row):
    if pd.isnull(row['Age']):
        return age_means[row['Pclass']]
    return row['Age']

titanicDisaster['Age'] = titanicDisaster.apply(fill_age, axis=1)

# Xử lý Cabin bị thiếu bằng "Unknown" và tách loại Cabin (typeCabin)
titanicDisaster['Cabin'] = titanicDisaster['Cabin'].fillna('Unknown')
titanicDisaster['typeCabin'] = titanicDisaster['Cabin'].astype(str).str[0]

# Tạo đặc trưng familySize
titanicDisaster['familySize'] = 1 + titanicDisaster['SibSp'] + titanicDisaster['Parch']

# Tạo đặc trưng Alone
titanicDisaster['Alone'] = np.where(titanicDisaster['familySize'] == 1, 1, 0)

# Xem kết quả sau khi làm sạch
print("\nDữ liệu sau khi được làm sạch:")
print(titanicDisaster.head())

# --- KIỂM TRA CÂN BẰNG DỮ LIỆU ---

# Đếm số lượng thực tế của mỗi nhãn trong cột Survived
#print("Số lượng hành khách sống sót (1) và thiệt mạng (0):")
#print(titanicDisaster['Survived'].value_counts())

# Xem tỷ lệ phần trăm để đánh giá mức độ chênh lệch
#print("\nTỷ lệ phần trăm (%):")
#print(titanicDisaster['Survived'].value_counts(normalize=True) * 100)

# Trực quan hóa bằng biểu đồ cột
#plt.figure(figsize=(6, 4))
#sns.countplot(x='Survived', data=titanicDisaster, palette='Set1')
#plt.title('Biểu đồ kiểm tra cân bằng dữ liệu - Cột Survived')
#plt.xlabel('0 = Thiệt mạng, 1 = Sống sót')
#plt.ylabel('Số lượng hành khách')
#plt.show()
```
